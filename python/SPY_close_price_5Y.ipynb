```python
import pandas as pd
df = pd.read_csv("SPY_close_price_5Y.csv")
```

```python
df.head()
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort by date (important)
df = df.sort_values(by='Date')

# Calculate moving averages
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_200'] = df['Close'].rolling(window=200).mean()

# Identify Golden Cross
df['Signal'] = 0

# When 50-day MA crosses above 200-day MA
df['Signal'] = (df['MA_50'] > df['MA_200']).astype(int)

# Detect crossover points (Golden Cross)
df['Golden_Cross'] = df['Signal'].diff()

# Filter only Golden Cross events (1 = crossover up)
golden_crosses = df[df['Golden_Cross'] == 1]

# Display results
print("Golden Cross Events:")
print(golden_crosses[['Date', 'Close', 'MA_50', 'MA_200']])

# Save results to CSV (optional)
golden_crosses.to_csv('golden_cross_results.csv', index=False)
```
